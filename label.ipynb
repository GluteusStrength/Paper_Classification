```python
# 전체 저널에서 labeling한 저널에 있는 학회지들만 추려낸 list가 in_label에 들어있게 된다
in_label = [i for i in list(set(journals)) if i in list(set(label["journal"]))]
mk_data = []
# 학회지 디렉터리까지 만들어두기
def mk_pth(i):
    pth = os.path.join(path, i)
    mk_data.append(pth)
for i in in_label:
    mk_pth(i)
# mk_data에는 추려낸 학회지를 포함한 디렉터리가 들어있게 된다.
mk_data
```

```python
# JSON file을 읽어들이는 과정
def load_json(pth):
    # mk_data의 경로의 하위 디렉터리는 연도 정보
    year = os.listdir(pth)
    year = year[-1:]
    # 1년치의 정보를 추려낸다.
    under_dir = [os.path.join(pth, y) for y in year]
    for under in tqdm(under_dir):
        file_names = os.listdir(under)
        for name in file_names:
            files = os.path.join(under, name)
            journal, title, txt, keyword = None, None, None, None
            with open(files, "r", encoding = "utf-8") as f:
                DATA = json.load(f)
            KEYS = list(DATA.keys())
            txt = []
            if "journal" in KEYS:
                if "ko" in list(DATA["journal"].keys()):
                    journal = DATA["journal"]["ko"]
                    if journal != pth[23:]:
                        journal = pth[23:]
            if "title" in KEYS:
                if "ko" in list(DATA["title"].keys()):
                    title = DATA["title"]["ko"]
            if "body_text" in KEYS:
                for t in DATA["body_text"]:
                    if "text" in list(t.keys()):
                        txt.append(t["text"][0])
                txt = " ".join(txt)
            if "keywords" in KEYS:
                if "ko" in list(DATA["keywords"].keys()):
                    keyword = DATA["keywords"]["ko"]
            data_list.append([journal, title, txt, keyword])
            

data_list = []
for pth in mk_data:
    load_json(pth)
```

```python
# 유효하지 않은 문장에 대해서 체크할 것.
# test set 역시도 이를 거치고 학습을 진행하도록 한다.
def check_right(data):
    for i in tqdm(range(len(data))):
        sentence = data.loc[i, "PAPER_TEXT"]
        f_sentence = []
        for sent in sentence.split("."):
            if len(sent) >= 20:
                f_sentence.append(sent.strip())
        data.loc[i, "PAPER_TEXT"] = ". ".join(f_sentence)

check_right(data)
```

```python
# 중분류 체계를 우선 만들기
target1 = list(data["RCMN_CD1"])
target2 = list(data["RCMN_CD2"])
target3 = list(data["RCMN_CD3"])
middle = []
for i in range(len(target1)):
    tg1, tg2, tg3 = target1[i], target2[i], target3[i]
    if not pd.isna(tg1):
        middle.append(tg1[:4])
    if not pd.isna(tg2):
        middle.append(tg2[:4])
    if not pd.isna(tg3):
        middle.append(tg3[:4])

middle_dict = dict()
middle = sorted(list(set(middle)))
for i in range(len(middle)):
    middle_dict[middle[i]] = i
```

```python
"""
만약 target1, target2, target3이 모두 존재하면 text를 3개로 분리
2개가 존재하면 text를 2개로 분리
1개만 존재하면 text 분리는 없다.
"""
# data = train
def split_data():
    DATA = []
    for i in tqdm(range(len(data))):
        target1, target2, target3 = data.loc[i, "RCMN_CD1"], data.loc[i, "RCMN_CD2"], data.loc[i, "RCMN_CD3"]
        if not pd.isna(target1) and not pd.isna(target2) and not pd.isna(target3):
            sentence = data.loc[i, "PAPER_TEXT"]
            sentence = sentence.split(".")
            length = len(sentence)
            sentence1 = sentence[:length // 3]
            sentence2 = sentence[length // 3: (length * 2) // 3]
            sentence3 = sentence[(length*2) // 3:]
            DATA.append([".".join(sentence1), target1])
            DATA.append([".".join(sentence2), target2])
            DATA.append([".".join(sentence3), target3])
        if not pd.isna(target1) and not pd.isna(target2) and pd.isna(target3):
            sentence = data.loc[i, "PAPER_TEXT"]
            sentence = sentence.split(".")
            length = len(sentence)
            sentence1 = sentence[:length//2]
            sentence2 = sentence[length//2:]
            DATA.append([".".join(sentence1), target1])
            DATA.append([".".join(sentence2), target2])
        if not pd.isna(target1) and not pd.isna(target3) and pd.isna(target2):
            sentence = data.loc[i, "PAPER_TEXT"]
            sentence = sentence.split(".")
            length = len(sentence)
            sentence1 = sentence[:length//2]
            sentence2 = sentence[length//2:]
            DATA.append([".".join(sentence1), target1])
            DATA.append([".".join(sentence2), target3])
        if not pd.isna(target1) and pd.isna(target3) and pd.isna(target2):
            continue
    return DATA
mk_data = split_data()
```
